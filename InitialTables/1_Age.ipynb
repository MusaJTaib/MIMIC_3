```python
#List of first admissions
HADM = np.load('MIMIC_3/admission_first_ids.npy',allow_pickle=True)
HADM = HADM.item().get('admission_ids')
```

```python
ADMISSIONS = pd.read_parquet("MIMIC_3/ADMISSIONS.parquet")
PATIENTS = pd.read_parquet("MIMIC_3/PATIENTS.parquet")
```

```python
X = ADMISSIONS['hadm_id'].isin(HADM)
```

```python
ADMISSIONS_df = ADMISSIONS[X]
```

```python
DOB1 = PATIENTS_df[['subject_id','dob']]
CurrentYear1 = ADMISSIONS_df[['subject_id','admittime']]

CurrentYear = CurrentYear1.copy(deep=True)
DOB = DOB1.copy(deep=True)

DOB['dob'] = [d.date() for d in DOB1['dob']]
CurrentYear['Current_date'] = [d.date() for d in CurrentYear1['admittime']]
CurrentYear['Current_time'] = [d.time() for d in CurrentYear1['admittime']]

CurrentYear = CurrentYear.sort_values(by=['subject_id'], ascending=True)
DOB = DOB.sort_values(by=['subject_id'], ascending=True)

DOB.reset_index(drop=True,inplace=True)
CurrentYear.reset_index(drop=True,inplace=True)
```

```python
CurrentYear.drop_duplicates(subset ="subject_id", keep = 'first', inplace = True)

DOB.drop_duplicates(subset ="subject_id", keep = 'first', inplace = True)
```

```python
X = CurrentYear['subject_id']
X.reset_index(drop=True,inplace=True)

Check = DOB['subject_id'].isin(X)
Check.unique()
DOB = DOB[Check]

DOB.reset_index(drop=True,inplace=True)
CurrentYear.reset_index(drop=True,inplace=True)

CurrentYear = CurrentYear.drop(['admittime','Current_time'], axis=1)
```

```python
import datetime as dt

DOB['dob'] = pd.to_datetime(DOB.dob, format='%Y-%m-%d')
CurrentYear['Current_date'] = pd.to_datetime(CurrentYear.Current_date, format='%Y-%m-%d')
```

```python
#days_in_year = 365.2425]
import numpy as np
df = pd.DataFrame()
df['subject_id'] = DOB['subject_id']
df['CurrentDate'] = CurrentYear['Current_date'] 
df['BirthDate'] = DOB['dob']

#df['Age'] = df.apply(lambda e: (e['CurrentDate'] - e['BirthDate']).days/365, axis=1)

df['age']  = ((df['CurrentDate'].values  - df['BirthDate'].values).astype(np.int)/8.64e13//365).astype(np.int)
```

```python
df = df.drop(df.index[df['age'] < 1])
df.reset_index(drop=True,inplace=True)
```
